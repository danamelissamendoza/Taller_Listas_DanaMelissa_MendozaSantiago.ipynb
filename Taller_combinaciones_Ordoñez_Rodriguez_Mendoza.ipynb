```python
import unicodedata

def normalizar(texto):

    texto = texto.replace('\xa0', ' ').strip()
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    return texto.lower()

def encontrar_archivo(ruta_carpeta, patron):

    patron_normalizado = normalizar(patron)
    for archivo in os.listdir(ruta_carpeta):
        if patron_normalizado in normalizar(archivo):
            return os.path.join(ruta_carpeta, archivo)
    return None

# Las llaves que identifican a cada persona en la GEIH
LLAVES_PERSONA = ['DIRECTORIO', 'SECUENCIA_P', 'ORDEN']
LLAVES_HOGAR   = ['DIRECTORIO', 'SECUENCIA_P']

# Parámetros comunes para todos los read_csv
PARAMS_CSV = {
    'sep'       : ';',
    'encoding'  : 'latin-1',
    'low_memory': False
}

# Patrones sin tildes para evitar problemas de codificación
PATRON_A = 'Caracteristicas generales'
PATRON_B = 'hogar y la vivienda'
PATRON_C = 'Fuerza de trabajo'
PATRON_D = 'Ocupados'

# Verificación rápida en Julio
ruta_julio = os.path.join(path, 'JULIO 2025')
print("🔍 Prueba en Julio:")
print(f"  A: {encontrar_archivo(ruta_julio, PATRON_A)}")
print(f"  B: {encontrar_archivo(ruta_julio, PATRON_B)}")
print(f"  C: {encontrar_archivo(ruta_julio, PATRON_C)}")
print(f"  D: {encontrar_archivo(ruta_julio, PATRON_D)}")
```

```python
# Nombres exactos de carpetas tal como aparecen en Drive
meses = {
    'Julio'      : 'JULIO 2025',
    'Agosto'     : 'Agosto 2025',
    'Septiembre' : 'Septiembre 2025',
    'Octubre'    : 'octubre',
    'Noviembre'  : 'noviembre',
    'Diciembre'  : 'diciembre'
}

lista_bases = []

for etiqueta, nombre_carpeta in meses.items():

    ruta_mes = os.path.join(path, nombre_carpeta)


    if not os.path.exists(ruta_mes) or len(os.listdir(ruta_mes)) == 0:
        print(f"⚠️  {etiqueta}: carpeta vacía o no encontrada, se omite.\n")
        continue

    print(f"\n{'='*55}")
    print(f"Procesando: {etiqueta}")
    print(f"{'='*55}")


    ruta_a = encontrar_archivo(ruta_mes, PATRON_A)
    ruta_b = encontrar_archivo(ruta_mes, PATRON_B)
    ruta_c = encontrar_archivo(ruta_mes, PATRON_C)
    ruta_d = encontrar_archivo(ruta_mes, PATRON_D)


    faltantes = [n for n, r in [('A', ruta_a), ('B', ruta_b),
                                 ('C', ruta_c), ('D', ruta_d)] if r is None]
    if faltantes:
        print(f"Módulos no encontrados: {faltantes}, se omite el mes.")
        continue


    df_a = pd.read_csv(ruta_a, **PARAMS_CSV)
    df_b = pd.read_csv(ruta_b, **PARAMS_CSV)
    df_c = pd.read_csv(ruta_c, **PARAMS_CSV)
    df_d = pd.read_csv(ruta_d, **PARAMS_CSV)


    df_a.columns = df_a.columns.str.upper().str.strip()
    df_b.columns = df_b.columns.str.upper().str.strip()
    df_c.columns = df_c.columns.str.upper().str.strip()
    df_d.columns = df_d.columns.str.upper().str.strip()

    print(f"Módulo A (Características): {df_a.shape[0]:,} filas")
    print(f"Módulo B (Hogar):           {df_b.shape[0]:,} filas")
    print(f"Módulo C (Fuerza trabajo):  {df_c.shape[0]:,} filas")
    print(f"Módulo D (Ocupados):        {df_d.shape[0]:,} filas")


    base = pd.merge(df_a, df_b, on=LLAVES_HOGAR,  how='left', suffixes=('', '_drop'))
    base = pd.merge(base, df_c, on=LLAVES_PERSONA, how='left', suffixes=('', '_drop'))
    base = pd.merge(base, df_d, on=LLAVES_PERSONA, how='left', suffixes=('', '_drop'))


    cols_drop = [c for c in base.columns if c.endswith('_drop')]
    base.drop(columns=cols_drop, inplace=True)


    base['MES'] = etiqueta

    print(f"\n  🔗 Base unida: {base.shape[0]:,} personas × {base.shape[1]} variables")

    lista_bases.append(base)

print("\n✅ Procesamiento terminado.")
```

```python
df_geih = pd.concat(lista_bases, axis=0, ignore_index=True)

print(f"Base semestral GEIH 2025:")
print(f"   Total personas-mes: {df_geih.shape[0]:,}")
print(f"   Total variables:    {df_geih.shape[1]}")
```
